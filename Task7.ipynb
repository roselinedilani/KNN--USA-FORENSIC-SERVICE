```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from scipy.spatial import distance
```

```python
Training_sample = pd.read_csv("/content/trainKNN.txt",header = None)
Training_sample
```

```python
new_sample_train = pd.DataFrame(Training_sample)
new_sample_train
```

```python
new_sample_train.columns = ['ID', 'RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe', 'Type of glass']
new_sample_train
```

```python
new_sample_train  = new_sample_train.drop('ID',axis =1)
new_sample_train
```

```python
test_sample = pd.read_csv("/content/testKNN.txt",header = None)
test_sample
```

```python
new_test_sample = pd.DataFrame(test_sample)
new_test_sample
```

```python
new_test_sample.columns = ['ID', 'RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe', 'Type of glass']
new_test_sample
```

```python
new_test_sample = new_test_sample.drop("ID" , axis = 1)
new_test_sample
```

```python
new_sample_train.head()
```

```python
new_test_sample.head()
```

```python
new_sample_train.describe()
```

```python
new_test_sample.describe()
```

```python
def standardize_data (df):
    for col in df.columns:
        if col != "Type of glass": 
            df[col] = (df[col] - df[col].mean())/df[col].std()
    return df

new_sample_train = standardize_data(new_sample_train)
new_test_sample = standardize_data(new_test_sample )
```

```python
new_sample_train.head()
```

```python
new_test_sample.head()
```

```python
euclid_model = KNeighborsClassifier(n_neighbors=8, metric=distance.sqeuclidean) 
manhattan_model = KNeighborsClassifier(n_neighbors=8, metric=distance.cityblock) 
x_train = new_sample_train.drop(["Type of glass"], axis=1)
y_train = new_sample_train["Type of glass"]
euclid_model.fit(x_train,y_train) 
manhattan_model.fit(x_train, y_train)
x_test = new_test_sample.drop("Type of glass", axis=1) 
y_test = new_test_sample["Type of glass"]
```

```python
# Make predictions
manhattan_predictions = manhattan_model.predict(x_test)
euclid_predictions = euclid_model.predict(x_test) 
df = pd.DataFrame({'actual': y_test, 'manhattan': manhattan_predictions, 'euclid': euclid_predictions})
df.head()
```

```python
manhattan_count = len(df.loc[df['manhattan'] == df['actual']])
euclid_count = len(df.loc[df['euclid'] == df['actual']])
print('Manhattan Accuracy: {}%'.format(round(100*manhattan_count/len(df), 2)))
print(classification_report(y_test, manhattan_predictions, target_names=df['actual'].astype(str).unique()))
print ('\n')
print('Square Euclidean Accuracy: {}%'.format(round(100*euclid_count/len(df), 2)))
print(classification_report(y_test, euclid_predictions, target_names=df['actual'].astype(str).unique()))
```
